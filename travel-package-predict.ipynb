# DATA PRE-PROCESSING

```python
# df['Gender'].replace('Fe Male', 'Female', inplace=True)
```

```python
df['Age_Bin'] = pd.cut(df['Age'], bins=[18, 25, 30, 40, 50, 70], labels=['18-25', '26-30', '31-40', '41-50', '51-70'])
```

```python
df['Salary_Bin'] = pd.cut(df['MonthlyIncome'], bins=[0, 20000, 40000, 60000, 80000, 100000], labels=['<20000', '<40000', '<60000', '<80000', '<100000'])
```

```python
df.drop(['Age', 'MonthlyIncome'], axis=1, inplace=True)
```

## HANDLING MISSING VALUES

```python
df['DurationOfPitch'] = df['DurationOfPitch'].fillna(0)
df['NumberOfTrips'] = df['NumberOfTrips'].fillna(0)
df['NumberOfChildrenVisiting'] = df['NumberOfChildrenVisiting'].fillna(0)
df['NumberOfFollowups'] = df['NumberOfFollowups'].fillna(0)
```

```python
df['TypeofContact'].value_counts()
```

```python
df['TypeofContact'] = df['TypeofContact'].fillna('Self Enquiry')
```

```python
df['PreferredPropertyStar'].value_counts()
```

```python
df['PreferredPropertyStar'] = df['PreferredPropertyStar'].fillna(df['PreferredPropertyStar'].mode()[0])
```

```python
df.dropna(subset=['Age_Bin', 'Salary_Bin'], inplace=True)
df.reset_index(drop=True, inplace=True)
```

## CONVERTING CATEGORICAL VARIABLES TO NUMERIC DATA

```python
numeric_cols = df.describe().columns
categ_cols = df.drop(numeric_cols, axis=1).columns
categ_cols
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
df_encoded = encoder.fit_transform(df[categ_cols])
df_encoded = pd.DataFrame(df_encoded.toarray(), columns=encoder.get_feature_names_out())
df = pd.concat([df, df_encoded], axis=1)
df.drop(categ_cols, axis=1, inplace=True)
```

# TRAIN TEST SPLIT

```python
X = df.drop('ProdTaken', axis=1)
y = df['ProdTaken']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

```python
from sklearn.ensemble import BaggingClassifier

bag_clf = BaggingClassifier(n_estimators=100, random_state=0)
bag_clf.fit(X_train, y_train)
y_pred = bag_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier(n_estimators=100, random_state=0)
rf_clf.fit(X_train, y_train)
y_pred = rf_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

```python
from sklearn.ensemble import AdaBoostClassifier

ada_clf = AdaBoostClassifier(n_estimators=100, random_state=0)
ada_clf.fit(X_train, y_train)
y_pred = ada_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_clf = GradientBoostingClassifier(n_estimators=100, random_state=0)
gb_clf.fit(X_train, y_train)
y_pred = gb_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

```python
X_train = X_train.copy()
X_train.columns = X_train.columns.str.replace('<', '')

X_test = X_test.copy()
X_test.columns = X_test.columns.str.replace('<', '')

from xgboost import XGBClassifier

xgb_clf = XGBClassifier(n_estimators=100, random_state=0)
xgb_clf.fit(X_train, y_train)
y_pred = xgb_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```
